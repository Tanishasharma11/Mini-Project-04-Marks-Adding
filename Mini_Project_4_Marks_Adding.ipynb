```python
import csv

def merge_marks(input_file):
    roll_marks = {}
    
    # Read input file and merge marks
    with open(input_file, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header row
        
        for row in reader:
            roll_number = row[0]
            marks = int(row[1])
            
            if roll_number in roll_marks:
                roll_marks[roll_number] += marks
            else:
                roll_marks[roll_number] = marks
    
    # Generate output file name
    output_file = 'output.csv'
    


    # Write output file with unique roll numbers and aggregated marks
    with open(output_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Roll Number', 'Marks'])
        
        for roll_number, marks in roll_marks.items():
            writer.writerow([roll_number, marks])
    
    print("Output file created successfully.")

    # Display the merged roll numbers and aggregated marks
    print("Roll Number\tMarks")
    for roll_number, marks in roll_marks.items():
      print(f"{roll_number}\t{marks}")

# Specify the input file path
input_file = '/content/Mini Project - Marks Adding.csv'

# Call the function to merge marks and generate the output file
merge_marks(input_file)
```
